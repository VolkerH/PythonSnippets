## Quick demonstration of new `skimage.measure.expand_labels` functionality (proposed Pull Request)

```python
import skimage
import napari
from napari.utils import nbscreenshot
import skimage.measure
import numpy as np
%gui qt
```

## 2D demo

```python
coins = skimage.data.coins()
coinlabels = skimage.measure.label(coins>100)
# expand labels by up to 10 pixels
expanded = skimage.measure.expand_labels(coinlabels, 10)
```

```python
v2d = napari.view_image(coins)
v2d.add_labels(coinlabels, name="orig labels")
```

```python
nbscreenshot(v2d)
```

```python
v2d.add_labels(expanded, name="expand_labels")
nbscreenshot(v2d)
```

# 3D example

```python
label_volume =  skimage.measure.label(np.random.randint(0,100, (100,100,100)) > 97)
```

```python
v3d = napari.viewer.Viewer()
v3d.add_labels(label_volume, name="seeds")
```

```python
nbscreenshot(v3d)
```

```python
expanded3d = skimage.measure.expand_labels(label_volume, 5)
```

```python
v3d.add_labels(expanded3d, name="expanded 3d labels")
```

```python
nbscreenshot(v3d)
```

### Show matrix values and greyvalue using pandas

This is just a snippet that demonstrates how to show a matrix with values using pandas with
each value backed by a greylevel representation. 
Note that this will not render properly in the Gitub/Gitlab notebook preview.

This might be quite useful for image analysis courses to demonstrate how an image is a matrix of numbers.

The idea is from one of the fastai course notebooks. Just capturing it here for future reference.

```python
import skimage.data
import pandas as pd
from skimage.transform import downscale_local_mean
```

```python
img = skimage.data.camera()
img.shape
```

```python
small_img = downscale_local_mean(img, (16, 16)).astype(int)
small_img.shape
```

```python
df = pd.DataFrame(small_img)
df.style.set_properties(**{'font-size':'7pt'}).background_gradient('Greys')
```
